```python
# mã sinh input cải tiến
import numpy as np
import json
import os
import math

def generate_nodes(N, L=400, energy_node=100, filename="nodes.json"):
    """
    Sinh N nút được phân bố đều và ngẫu nhiên trong không gian 3D
    Mỗi nút nằm ngẫu nhiên trong 1 khối lập phương con (sub-cube).
    """
    np.random.seed(0)  # để tái lập kết quả
    
    # Tính số khối theo mỗi trục gần bằng căn bậc 3 của N
    n_per_dim = math.ceil(N ** (1/3))
    cube_size = L / n_per_dim  # kích thước 1 khối lập phương con
    
    data = []
    node_id = 0

    # Duyệt qua từng khối lập phương con
    for i in range(n_per_dim):
        for j in range(n_per_dim):
            for k in range(n_per_dim):
                if node_id >= N:
                    break
                # Giới hạn tọa độ của khối lập phương con
                x_min, x_max = i * cube_size, (i + 1) * cube_size
                y_min, y_max = j * cube_size, (j + 1) * cube_size
                z_min, z_max = k * cube_size, (k + 1) * cube_size

                # Đặt nút ngẫu nhiên trong khối này
                x = np.random.uniform(x_min, x_max)
                y = np.random.uniform(y_min, y_max)
                z = np.random.uniform(z_min, z_max)

                data.append({
                    "id": node_id,
                    "x": round(float(x), 3),
                    "y": round(float(y), 3),
                    "z": round(float(z), 3),
                    "energy_residual" : float(energy_node),
                    "energy_node": float(energy_node)
                })

                node_id += 1
            if node_id >= N:
                break
        if node_id >= N:
            break

    os.makedirs("D:/Year 4/tiến hóa/project/data/input_data_evenly_distributed", exist_ok=True)
    filepath = f"D:/Year 4/tiến hóa/project/data/input_data_evenly_distributed/{filename}"
    with open(filepath, "w") as f:
        json.dump(data, f, indent=4)

    print(f"Tạo file {filepath} chứa {N} node (phân bố đều và ngẫu nhiên).")

# Sinh bộ dữ liệu thử nghiệm
for N in [150, 200, 250, 300, 350, 400, 450, 500, 550]:
    generate_nodes(N, L=400, filename=f"nodes_{N}.json")
```
